```r
# view the data and create a new object
View(glass)
```

```r
ConeBeakers <- glass %>%
  filter(str_detect(Form, "Cone beaker"))

## Different vessel form
Bowl <- glass %>%
  filter(str_detect(Form, "Bowl"))
```

```r
# double check you've got the stuff you want, eg:
ConeBeakers
Bowl
```

```r
# Make a bar plot of how often each site appears in the Cone Beaker table.
# first we use the 'table' command to count up the number of times each site appears
siteCounts <- table(ConeBeakers$Site)

# then we sort the list
siteCounts <- sort(siteCounts, decreasing=TRUE)

# check
siteCounts
```

```r
#barplot(data, title, label, show labels, make 'em really small)
barplot(siteCounts, main="Sites", xlab="Site", las=2, cex.names=.5 )
```

```r
# Wanted to also see it for our different vessel we've chosen
siteCounts1 <- table(Bowl$Site)

# then we sort the list
siteCounts <- sort(siteCounts1, decreasing=TRUE)

# check
siteCounts1

## bar plot
barplot(siteCounts1, main="Sites", xlab="Site", las=2, cex.names=.5 )
```

```r
# We can follow the same pattern as when we searched for cone beakers,
# but searching the 'site' column instead from our original 'glass' data.

Whithorn <- glass %>%
  filter(str_detect(Site, "WHITHORN"))

View(Whithorn)
```

```r
# Now we make a barplot for the other kinds of forms at that site

whithornCounts <- table(Whithorn$Form)
whithornCounts <- sort(whithornCounts, decreasing=TRUE)
barplot(whithornCounts, main="Fomrs", xlab="Form", las=2, cex.names=.5)

crosstab <- xtabs(~Form+Group, Whithorn)
crosstab
barplot(crosstab, las=2)
```

```r
# Basic R code taught before in the course
# from week 5
library(curl)
library(stringr)
library(dplyr)

# Load the glass data from ADS
glass <- read.csv(curl("https://archaeologydataservice.ac.uk/archiveDS/archiveDownload?t=arch-788-1/dissemination/csv/imports_database/Glass.csv"), header = TRUE, row.names="ID")

# View the data structure
View(glass)
summary(glass)
str(glass)
```

```r
# Chi-square 
chi_test <- chisq.test(glass$Form, glass$Group)
print(chi_test)

# when running this, we can conclude that strong evidence that 
# certain glass forms are statistically associated with specific time periods
```

```r
# Most common glass forms in the dataset and how often do they appear
form_counts <- table(glass$Form)
form_counts_sorted <- sort(form_counts, decreasing = TRUE)
print(form_counts_sorted)
```

```r
# Bar plot 
barplot(form_counts_sorted[1:15], 
        main = "Most Common Glass Forms",
        las = 2,
        cex.names = 0.7,
        ylab = "Frequency")
```

```r
# Which sites have the most glass finds? We know that Whithorn appeared the most in the Cone Beaker table but outside of it

site_counts <- table(glass$Site)
site_counts_sorted <- sort(site_counts, decreasing = TRUE)
print(site_counts_sorted)
```
